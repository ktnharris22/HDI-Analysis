```python
import pandas as pd
import json

hdi_df = pd.read_csv("HDI.csv",header=5,nrows=189)
hdi_df.dropna(axis='columns', inplace=True)
display(hdi_df)

with open('factbook.json',encoding="utf8") as f:
  factbook = json.load(f)

factbook = factbook.get("countries")
print(len(factbook.keys()))
factbook.keys()

pop = factbook.pop("world")
pop = factbook.pop("european_union")
pop = factbook.pop("arctic_ocean")
pop = factbook.pop("atlantic_ocean")
pop = factbook.pop("british_indian_ocean_territory")
pop = factbook.pop("pacific_ocean")
pop = factbook.pop("western_sahara")

print(len(factbook.keys()))
hdi_df["Country"] = hdi_df["Country"].str.lstrip()
print(list(hdi_df["Country"]))
hdi_2019=hdi_df["2019"]
data=[hdi_df["Country"], hdi_df["2019"]]
hdi_df_2019 = pd.concat(data, axis=1, keys=["Country","2019"])
hdi_df_2019.rename(columns={'2019':'HDI_2019'}, inplace=True)

display(hdi_df_2019)
#http://hdr.undp.org/en/indicators/137506
```

```python
import matplotlib.pyplot as plt

hdi_df.drop(["HDI Rank"], axis=1, inplace=True)

print(hdi_df.columns)
display(hdi_df)
```

```python
from IPython.display import display
import datetime as dt
import numpy as np
tday = dt.datetime.now()

cols = ["Name", "Lat", "Long", "Map_Ref", "Area", "Population", "Language", "Religion", "Median_age", "pop_growth_rate", "birth_rate", "death_rate", "net_migration_rate", "sex_ratio", "maternal_mortality_rate", "mothers_mean_age_at_first_birth", "infant_mortality_rate","life_exp","physician_density","drinking_water_source","adult_obesity", "education_exp", "literacy_rate","government_type","gdp_rank","real_growth_rate","primary_sector","gross_national_savings","labor_force","unemployment_rate","total_electrification","num_airports","paved_road","unpaved_road","military_exp","day_since_indp"]

factbook_df = pd.DataFrame(columns=cols)

for country in factbook.keys():
    c = factbook[country]["data"]
    Name = c["name"]
    if Name in list(hdi_df["Country"]):
        print(Name)
        if "geography" in c.keys():
            c_geo = c["geography"]
            lat_dict = c_geo["geographic_coordinates"]["latitude"]
            long_dict = c_geo["geographic_coordinates"]["longitude"]
            Lat = float(str(lat_dict["degrees"]) + "." + str(lat_dict["minutes"]))
            Lat = Lat if lat_dict["hemisphere"] =='N' else Lat*-1
            Long = float(str(long_dict["degrees"]) + "." + str(long_dict["minutes"]))
            Long = Long if long_dict["hemisphere"] == "E" else Long*-1

            Map_ref = c_geo["map_references"]
            Area = c_geo["area"]["total"]["value"]
        else:
            Lat = np.nan
            Long = np.nan
            Map_ref = np.nan
            Area = np.nan
 
        if "people" in c.keys():
            c_peo = c["people"]
            Population = c_peo["population"]["total"]
            Lang = c_peo["languages"]["language"][0]["name"]
            Rel = c_peo["religions"]["religion"][0]["name"] if "religions" in c_peo.keys() else np.nan
            Med_age = c_peo["median_age"]["total"]["value"] if "median_age" in c_peo.keys() else np.nan
            pop_growth_rate = c_peo["population_growth_rate"]["growth_rate"] if "population_growth_rate" in c_peo.keys() else np.nan
            birth_rate = c_peo["birth_rate"]["births_per_1000_population"] if "birth_rate" in c_peo.keys() else np.nan
            death_rate = c_peo["death_rate"]["deaths_per_1000_population"] if "death_rate" in c_peo.keys() else np.nan
            net_migration_rate = c_peo["net_migration_rate"]["migrants_per_1000_population"] if "net_migration_rate" in c_peo.keys() else np.nan
            sex_ratio = c_peo["sex_ratio"]["by_age"]["at_birth"]["value"] if "sex_ratio" in c_peo.keys() else np.nan
            mothers_mean_age_at_first_birth = c_peo["mothers_mean_age_at_first_birth"]["age"] if "mothers_mean_age_at_first_birth" in c_peo.keys() else np.nan
            maternal_mortality_rate = c_peo["maternal_mortality_rate"]["deaths_per_100k_live_births"] if "maternal_mortality_rate" in c_peo.keys() else np.nan
            infant_mortality_rate = c_peo["infant_mortality_rate"]["total"]["value"] if "infant_mortality_rate" in c_peo.keys() else np.nan
            life_exp = c_peo["life_expectancy_at_birth"]["total_population"]["value"] if "life_expectancy_at_birth" in c_peo.keys() else np.nan
            physician_density = c_peo["physicians_density"]["physicians_per_1000_population"] if "physicians_density" in c_peo.keys() else np.nan
            if "drinking_water_source" in c_peo.keys():
                if "improved" in c_peo["drinking_water_source"].keys():
                    drinking_water_source = c_peo["drinking_water_source"]["improved"]["total"]["value"] if "total" in c_peo["drinking_water_source"]["improved"].keys() else np.nan
            else:
                np.nan
            adult_obes = c_peo["adult_obesity"]["percent_of_adults"] if "adult_obesity" in c_peo.keys() else np.nan
            educ_exp = c_peo["education_expenditures"]["percent_of_gdp"] if "education_expenditures" in c_peo.keys() else np.nan
            literacy_rate = c_peo["literacy"]["total_population"]["value"] if "literacy" in c_peo.keys() else np.nan

        if "government" in c.keys():
            c_gov = c["government"]
            if "government_type" in c_gov.keys():
                gov_type = c_gov["government_type"]
            
            if "independence" in c_gov.keys():
                ind_date = c_gov["independence"]["date"] if "date" in c_gov["independence"].keys() else np.nan
                ind_from = c_gov["independence"]["note"] if "note" in c_gov["independence"].keys() else "N/A"
                if not pd.isna(ind_from):
                    ind_from = ind_from[5:]
                    if ';' in ind_from:
                        ind_from = ind_from[:ind_from.index(";")]
            else:
                ind_date = np.nan
                ind_from = "N/A"

            if pd.isna(ind_date):
                day_since_indp = 100000
            else:
                ind_date = dt.datetime.strptime(ind_date, '%Y-%m-%d')
                day_since_indp = (tday - ind_date).days

        if "economy" in c.keys():
            c_eco = c["economy"]
            if "gdp" in c_eco.keys():
                gdp_rank = c_eco["gdp"]["purchasing_power_parity"]["global_rank"]
                real_growth_rate = c_eco["gdp"]["real_growth_rate"]["annual_values"][0]["value"]
                if "by_sector_of_origin" in c_eco["gdp"]["composition"].keys():
                    sectors = c_eco["gdp"]["composition"]["by_sector_of_origin"]["sectors"]
                    max_s = 0
                    for s in sectors.keys():
                        if(sectors[s]["value"] > max_s):
                            max_s = sectors[s]["value"]
                            primary_sector = s
                else:
                    primary_sector = "N/A"
            else:
                gdp_rank = np.nan
                real_growth_rate = np.nan
                primary_sector = "N/A"

            gross_nat_sav = c_eco["gross_national_saving"]["annual_values"][0]["value"] if "gross_national_saving" in c_eco.keys() else np.nan
            labor_force = c_eco["labor_force"]["total_size"]["total_people"] if "labor_force" in c_eco.keys() else np.nan
            
            if "unemployment_rate" in c_eco.keys():
                unemployment_rate = c_eco["unemployment_rate"]["annual_values"][0]["value"] if "annual_values" in c_eco["unemployment_rate"].keys() else np.nan
            else:
                unemployment_rate = np.nan
        
        if "energy" in c.keys():
            if "electricity" in c["energy"].keys():
                total_electrification = c["energy"]["electricity"]["access"]["total_electrification"]["value"] if "access" in c["energy"]["electricity"].keys() else np.nan
            else:
                total_electrification = np.nan

        if "transportation" in c.keys():
            if "air_transport" in c["transportation"].keys():
                num_airports = c["transportation"]["air_transport"]["airports"]["total"]["airports"] if "airports" in c["transportation"]["air_transport"].keys() else np.nan
            if "roadways" in c["transportation"].keys():
                paved_road = c["transportation"]["roadways"]["paved"]["value"] if "paved" in c["transportation"]["roadways"].keys() else np.nan
                unpaved_road = c["transportation"]["roadways"]["unpaved"]["value"] if "unpaved" in c["transportation"]["roadways"].keys() else np.nan
            else:
                paved_road = np.nan
                unpaved_road = np.nan

        if "military_and_security" in c.keys():
            if("expenditures" in c["military_and_security"].keys()):
                Mil_exp = c["military_and_security"]["expenditures"]["annual_values"][0]["value"]
            else:
                Mil_exp = np.nan
                
        row_dict = {"Name":Name, "Lat":Lat , "Long":Long, "Map_Ref":Map_ref, "Area":Area, "Population":Population, "Language":Lang, "Religion":Rel, "Median_age":Med_age, "pop_growth_rate":pop_growth_rate, "birth_rate":birth_rate, "death_rate":death_rate, "net_migration_rate":net_migration_rate, "sex_ratio":sex_ratio, "maternal_mortality_rate":maternal_mortality_rate, "mothers_mean_age_at_first_birth":mothers_mean_age_at_first_birth, "infant_mortality_rate":infant_mortality_rate,"life_exp":life_exp,"physician_density":physician_density,"drinking_water_source":drinking_water_source,"adult_obesity":adult_obes, "education_exp":educ_exp, "literacy_rate":literacy_rate,"government_type":gov_type,"Independence_date":ind_date, "day_since_indp":day_since_indp, "Independence_from":ind_from, "gdp_rank":gdp_rank, "real_growth_rate":real_growth_rate,"primary_sector":primary_sector, "gross_national_savings":gross_nat_sav, "labor_force":labor_force, "unemployment_rate":unemployment_rate, "total_electrification":total_electrification,"num_airports":num_airports, "paved_road":paved_road, "unpaved_road":unpaved_road, "military_exp":Mil_exp}

        factbook_df = factbook_df.append(row_dict,ignore_index=True)

display(factbook_df)
```

```python
#MERGE 
merge_df = hdi_df[['Country',"2019"]].merge(factbook_df, left_on='Country', right_on='Name',how='inner')
merge_df.set_index("Country", inplace=True)
merge_df.rename(columns={'2019':'HDI_2019'}, inplace=True)
merge_df.drop(["Name", "Independence_date"], axis=1, inplace=True) # drop ind_date, replaced by days_since_indp
display(merge_df)
merge_df.columns
```

```python
merge_df = merge_df.convert_dtypes()
merge_df.dtypes
#merge_df["Independence_from"].unique()
```

```python
##Numeric Data Only
numeric_df = merge_df.select_dtypes(include=['Int64','float64'])
print(numeric_df.shape)

## Correlation Matrix
corr_matrix = numeric_df.corr()
from seaborn import heatmap
import matplotlib.pyplot as pyplot

pyplot.figure(figsize=(15,15))
heatmap(corr_matrix.apply(lambda x: round(x,2)), annot=True, square=True)
pyplot.show()
pyplot.savefig("corr_matrix.png")
```

```python
## Change Categorial Vars to encodings
encocded_df = pd.get_dummies(merge_df, columns=['Map_Ref','Language','Religion','government_type','primary_sector','Independence_from'], prefix=['continent','lang','rel','gov','pri_sector','ind_from'])
print(encocded_df.shape)
```

```python
HDI_corr = numeric_df.corr().iloc[0,].sort_values(ascending=False)
#display(HDI_corr.index)
plt.bar(height=HDI_corr, x=HDI_corr.index)
plt.xticks(rotation='vertical')
plt.title("Most Important Factors for HDI_2019")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
## Split Train and Test
X_encoded = encocded_df.drop(columns='HDI_2019',axis = 1)
Y_encoded = encocded_df["HDI_2019"]

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y_encoded, test_size=0.3, random_state=4)

#Calculate Adjusted R2
def adjR2(r2):
    n=X_encoded.shape[0]
    k=31
    return 1-(1-r2)*(n-1)/(n-k-1)

#Building and Testing Linear Models
from sklearn import linear_model as lin_mod
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics

#Basic Linear Regression with all numeric X vars
print("Basic Linear Model with 5 predictors included")
print("-"*100)
lm = lin_mod.LinearRegression()
X_train_select = X_train[["total_electrification","mothers_mean_age_at_first_birth","drinking_water_source","physician_density","literacy_rate","pop_growth_rate"]]
X_test_select =  X_test[["total_electrification","mothers_mean_age_at_first_birth","drinking_water_source","physician_density","literacy_rate","pop_growth_rate"]]
lm.fit(X_train_select,Y_train)
print(f"R-squared (training data) is {adjR2(lm.score(X_train_select,Y_train))}")
preds = lm.predict(X_test_select)
print(f"R-squared (testing) is {adjR2(metrics.r2_score(y_true=Y_test, y_pred = preds))}\n")

print("Basic Lasso Model with all predictors included")
print("-"*100)
lasso = lin_mod.Lasso(alpha = 0.0001)
lasso.fit(X_train, Y_train)
print(f"R-squared (training data) is {adjR2(lasso.score(X_train, Y_train))}")
preds = lasso.predict(X_test)
print(f"R-squared score (testing) is {adjR2(metrics.r2_score(y_true=Y_test, y_pred = preds))}\n")

print("Basic RF Model with all predictors included")
print("-"*100)
rf_regr = RandomForestRegressor(max_depth=10, random_state=0)
rf_regr.fit(X_train, Y_train)
print(f"R-squared (training data) is {adjR2(rf_regr.score(X_train, Y_train))}")
preds = rf_regr.predict(X_test)
print(f"R-squared score (testing) is {adjR2(metrics.r2_score(y_true=Y_test, y_pred = preds))}\n")

print("Basic Gradient Boosting Model with all predictors included")
print("-"*100)
params = {'n_estimators': 500,
          'max_depth': 4,
          'min_samples_split': 5,
          'learning_rate': 0.01,
          'loss': 'ls'}
gb_regr = GradientBoostingRegressor(**params)
gb_regr.fit(X_train, Y_train)
print(f"R-squared (training data) is {adjR2(gb_regr.score(X_train, Y_train))}")
preds = gb_regr.predict(X_test)
print(f"R-squared score (testing) is {adjR2(metrics.r2_score(y_true=Y_test, y_pred = preds))}\n")

```

```python
importance = np.abs(lasso.coef_)
feature_names = np.array(X_train.columns)


data = {'Name': feature_names, 'Imp': importance}
df = pd.DataFrame(data)
df = df.sort_values(by='Imp', ascending=False)
df_10 = df.head(10)

plt.bar(height=df_10['Imp'], x=df_10['Name'])
plt.xticks(rotation='vertical')
plt.title("Feature importances via coefficients (Lasso)")
plt.show()
```
